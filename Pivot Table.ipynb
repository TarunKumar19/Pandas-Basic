```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", np.nan,
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
pd.pivot_table(df,values=['D','E'] , index=['A'] ,columns='C' ,aggfunc=np.sum)
```

```python
table=pd.pivot_table(df,values=['D','E'] ,index=['A'],columns='C',aggfunc=np.sum ,dropna=False)
table
```
